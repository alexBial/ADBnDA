```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.special import logit, expit

#Exercise1 DataFrame of normal distribution with different parameters (loc, scale and size)
normal_dis = {
    "def":np.random.normal(size=10000),
    "10_1":np.random.normal(loc=10,scale=1,size=10000),
    "1_10":np.random.normal(loc=1,scale=10,size=10000)
}

df = pd.DataFrame(data=normal_dis)

df_melt = pd.melt(df,var_name='parameters',value_name='value')

g = sns.FacetGrid(df_melt,col="parameters",sharex=True)
g.map(plt.hist,'value',cumulative=False,bins=64)

g = sns.FacetGrid(df_melt,col="parameters",sharex=True)
g.map(plt.hist,'value',cumulative=True,bins=64)

# Value of bins determine estimation of graph, if value is bigger the sample is smaller, 
# we need to choose that value when resolution of graph will be ok without unexpected peaks
```

```python
#Exercise2 Inverse logit
# We use previous dataframe but this time it was transformed with expit function expit(x) = 1/(1+exp(-x))
expit_df = expit(df)
expit_df_melt = pd.melt(expit_df,var_name='parameters',value_name='value')

g = sns.FacetGrid(expit_df_melt,col="parameters",sharex=True)
g.map(plt.hist,'value',cumulative=False,bins=64)

g = sns.FacetGrid(expit_df_melt,col="parameters",sharex=True)
g.map(plt.hist,'value',cumulative=True,bins=16)
```

```python
#Exercise3 Poisson Disribution
# Standard deviation is equal of square root of expected value (Graph is getting more flat with increasing them)
poison_dis={
    "def":np.random.poisson(size=10000),
    "interval_5":np.random.poisson(lam=5,size=10000),
    "interval_10":np.random.poisson(lam=10,size=10000),
}

df_poison = pd.DataFrame(data=poison_dis)
df_poison_melt = pd.melt(df_poison,var_name='parameters',value_name='value')

g = sns.FacetGrid(df_poison_melt,col="parameters",sharex=True)
g.map(plt.hist,'value',cumulative=False,bins=64)

g = sns.FacetGrid(df_poison_melt,col="parameters",sharex=True)
g.map(plt.hist,'value',cumulative=False,bins=128)
```

```python
#Exercise4 Beta Distribution
# Beta distribution is continuous probability distribution defined on the interval (0,1). Shape depends on 2 parameters (alpha and beta)
beta_dis={
    "1_1":np.random.beta(a=1,b=1,size=10000),
    "10_11":np.random.beta(a=10,b=1,size=10000),
    "1_10":np.random.beta(a=1,b=10,size=10000), 
}

df_beta = pd.DataFrame(data=beta_dis)
df_beta_melt = pd.melt(df_beta,var_name='patameters',value_name='values')

g = sns.FacetGrid(df_beta_melt,col='patameters',sharex=True)
g.map(plt.hist,'values',cumulative=False,bins=64)
```

```python
#Exercise5 Log normal distribution
# It is only for positive values ( normal distribution also contain negative values). Shape depends on mean and sigma value
log_normal_dis={
    "default":np.random.lognormal(size=10000),
    "1_1":np.random.lognormal(mean=1,sigma=1,size=10000),
    "10_0.5":np.random.lognormal(mean=10,sigma=0.5,size=10000),
    "5_0.1":np.random.lognormal(mean=5,sigma=0.1,size=10000),    
}

df_log_normal = pd.DataFrame(data=log_normal_dis)

df_log_normal_melt = pd.melt(df_log_normal,var_name='patameters',value_name='values')

g = sns.FacetGrid(df_log_normal_melt,col='patameters',sharex=False,sharey=False)
g.map(plt.hist,'values',cumulative=False,bins=64)
```
