```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('C:\Data1.csv', index_col = 0)
df.index.name = 'dates'
df.head()
```

```python
ax = df.plot(figsize=(15,6))
df.plot(subplots= True,figsize=(15,6),legend=None)
plt.show()
```

```python
meltd_histg = pd.melt(df,id_vars=['theta_1'],value_vars=['theta_2','theta_3','theta_4','theta_5','theta_6'])
g = sns.FacetGrid(meltd_histg,sharex=False)
g = g.map(plt.hist,'value')
plt.show()
meltd_histg.head()
ax = df.plot.kde(figsize=(15,4))
```

```python
df2 = df.loc['2018-01-01':'2018-12-31','theta_1':'theta_4']
df2.head()
```

```python
ax = df2.plot(figsize = (15,6))
df2.plot(subplots=True,figsize=(15,6),legend = None)
plt.show()
```

```python
meltd_histg2 = pd.melt(df2,id_vars=['theta_1'],value_vars=['theta_2','theta_3','theta_4'])
g = sns.FacetGrid(meltd_histg2, sharex = False)
g = g.map(plt.hist, 'value')
plt.show()
meltd_histg2.head()
ax = df2.plot.kde(figsize=(15,4))
```
